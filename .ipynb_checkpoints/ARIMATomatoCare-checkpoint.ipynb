```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error
%matplotlib inline

trainingdata=pd.read_csv('ARIMAdataset.csv')
# Convert 'Year' column to string type
trainingdata['Year'] = trainingdata['Year'].astype(str)
# Create a new column combining 'Year' and 'TimePeriod'
trainingdata['YearQuarter'] = trainingdata['Year'] + '-' + trainingdata['TimePeriod']
```

```python
# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(trainingdata['YearQuarter'], trainingdata['AreaHarvested'])
plt.xlabel('Year-Quarter')
plt.ylabel('Area Harvested')
plt.title('Area Harvested by Year-Quarter')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()

# Plot ACF
plt.figure(figsize=(12, 6))
plot_acf(trainingdata['AreaHarvested'], ax=plt.gca(), lags=20)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function')
plt.show()

# Plot PACF
plt.figure(figsize=(12, 6))
plot_pacf(trainingdata['AreaHarvested'], ax=plt.gca(), lags=25)
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Function')
plt.show()

# ADF test
result = adfuller(trainingdata['AreaHarvested'])
print('ADF Statistic: ', result[0])
print('p-value: ', result[1])
print('Critical Values: ')
for key, value in result[4].items():
    print(f'{key}:{value}')
    
# Fit the ARIMA model
train_data = trainingdata['AreaHarvested'].iloc[:int(len(trainingdata) * 0.7)]
model = SARIMAX(train_data, order=(4, 1, 0), seasonal_order=(0, 1, 0, 4))
model_fit = model.fit()


# Make time series predictions for training set
train_forecast = model_fit.predict(start=0, end=int(len(train_data)-1))

# Make time series predictions
test_data = trainingdata['AreaHarvested'].iloc[int(len(trainingdata)* 0.7):]
forecast = model_fit.forecast(steps=len(test_data))

# Combine actual and forecast data for plotting
combined_data = pd.concat([trainingdata[['YearQuarter', 'AreaHarvested']], pd.Series(forecast)], axis=1)
combined_data.columns = ['YearQuarter', 'Actual', 'Forecast']
combined_data['AbsoluteError'] = np.abs(combined_data['Actual'] - combined_data['Forecast'])
absolute_percentage_error = abs((combined_data['Actual'] - combined_data['Forecast']) / combined_data['Actual']) * 100

# Evaluate model predictions
print('\nTraining Set:')
rmse_train = np.sqrt(mean_squared_error(train_data, train_forecast))
print('Root Mean Squared Error (RMSE): ', rmse_train)
mae_train = combined_data_train['AbsoluteError'].mean()
print('Mean Absolute Error (MAE):', mae_train)
mean_absolute_percentage_error_train = absolute_percentage_error_train.mean()
print('Mean Absolute Percentage Error (MAPE):', mean_absolute_percentage_error_train)

print('\nTest Set:')
rmse = np.sqrt(mean_squared_error(test_data, forecast))
print('Root Mean Squared Error (RMSE): ', rmse)
mae = combined_data['AbsoluteError'].mean()
print('Mean Absolute Error (MAE):', mae)
mean_absolute_percentage_error = absolute_percentage_error.mean()
print('Mean Absolute Percentage Error (MAPE):', mean_absolute_percentage_error)

# Plot the actual and forecasted data
plt.figure(figsize=(12, 6))
plt.plot(combined_data['YearQuarter'], combined_data['Actual'], label='Actual')
plt.plot(combined_data['YearQuarter'], combined_data['Forecast'], label='Forecast')
plt.xlabel('Year-Quarter')
plt.ylabel('Area Harvested')
plt.title('Actual vs Forecasted Data')
plt.legend()
plt.show()

# Forecasts future values
model = SARIMAX(trainingdata['AreaHarvested'], order=(4, 1, 0), seasonal_order=(0, 1, 0, 4))
model_fit = model.fit()


# prediction = model_fit.forecast(steps=8)
# print('Predicted Future Data')
# print(prediction)
# Make time series predictions for the future
future_steps = 4
forecast = model_fit.forecast(steps=future_steps)

# Get the last year and quarter from the training data
last_year = int(trainingdata['Year'].iloc[-1])
last_quarter = trainingdata['TimePeriod'].iloc[-1]

# Map quarter to numeric values
quarter_mapping = {'Q1': 1, 'Q2': 2, 'Q3': 3, 'Q4': 4}
last_quarter_numeric = quarter_mapping[last_quarter]

# Generate future year-quarters
future_year_quarters = []
for _ in range(future_steps):
    if last_quarter_numeric < 4:
        last_quarter_numeric += 1
    else:
        last_quarter_numeric = 1
        last_year += 1
    future_quarter = [key for key, value in quarter_mapping.items() if value == last_quarter_numeric][0]
    future_year_quarters.append(f"{last_year}-{future_quarter}")

# Plot the forecasted data
plt.figure(figsize=(12, 6))
plt.plot(trainingdata['YearQuarter'], trainingdata['AreaHarvested'], label='Historical Data')
plt.plot(future_year_quarters, forecast, label='Forecasted Data')
plt.xlabel('Year-Quarter')
plt.ylabel('Area Harvested')
plt.title('Actual and Forecasted Data')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()

print('Predicted Future Data:')
forecasted_data = list(forecast)  # Convert NumPy array to list
for i, year_quarter in enumerate(future_year_quarters):
    print(f'{year_quarter}: {forecasted_data[i]}')
```
